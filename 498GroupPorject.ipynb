# Importing libraries



```python
import pandas as pd
import matplotlib.pyplot as plt
```

# Reading in files



```python
# read in csv files
games = pd.read_csv('games.grivg.csv')
characters = pd.read_csv('characters.grivg.csv')
sexualization = pd.read_csv('sexualization.grivg.csv')
```

# Organizing the data

```python
# renaming column for merging purposes
characters = characters.rename(columns={'Game': 'Game_Id'})
```

```python
# merging the three datasets
VGame = pd.merge(characters, games, on='Game_Id', how='left')
VGame = pd.merge(VGame, sexualization, on='Id', how='left')
```

```python
# isolating the year in the release date so that we can analysis trends throughout the years
VGame['Year'] = VGame['Release'].str[-2:]
```

# What is the gender distribution in video game based on their genres?





```python
# Group by the genre then count for the gender
gender_by_genre = VGame.groupby('Genre')['Gender'].value_counts().unstack()

# Plotting the gender representation by genre
gender_by_genre.plot(kind='bar', figsize=(10, 6))
plt.title('Gender Representation in Video Games by Genre')
plt.xlabel('Genre')
plt.ylabel('Count')
plt.legend(title='Gender')

# Display the plot
plt.show()
```

This graph of "Gender Representation in Video Games by Genre" shows that Action and Action-adventure games tend to have a higher proportion of male characters compared to female characters. This finding aligns with the historical trend of male-dominated protagonists in these genres. Compare to Action and Action-adventure games, Adventure and RPGs demonstrate a more balanced gender representation, the number of female characters is more than half of the male characters. The Adventure game has the highest female proportion compared to others. I think it might be because the video game industry is slowly moving toward greater inclusivity and diversity in recent years. For RPGs, the genre also includes custom choice, indicating a conscious effort to provide diverse options for players. Through this graph, we could also see a lack of Interactive and Simulation type of video games in this dataset.

# What is the gender distribution in video game based on publishers?


```python
# groupby publisher for each gender catergory
gender_by_publisher = VGame.groupby('Publisher')['Gender'].value_counts().unstack()

# plotting the gender representation by publisher
gender_by_publisher.plot(kind='bar', stacked=True, figsize=(10, 6))

# adding title, labels, and legend to the graph
plt.title('Gender Representation vs Video Game Publishers')
plt.xlabel('Publisher')
plt.ylabel('Count')
plt.legend(title='Gender')

# display the plot
plt.show()
VGame["Publisher"].value_counts().count()
```

With this graph, we wanted to examine how each video game publisher did at gender representation when it comes to their most popular games. This graph reveals the proportion of male, female, and non-binary video game characters that each publisher represents. We can see that almost all of the publishers have more male characters in their video games than female characters. Additionally, only 18 out of 44 publishers have non-binary characters in their games.


# What is the gender distribution in video games from 2012-2022?


```python
# filtering through games and characters data
character_df = pd.merge(characters, games, on='Game_Id')
grouped_data = character_df.groupby(['Release', 'Gender']).size().unstack()
grouped_data = grouped_data.reset_index()
grouped_data['Release'] = grouped_data['Release'].str.extract(r'(\d+)') #remove month from release column
grouped_data = grouped_data.groupby('Release').agg({'Female': 'sum', 'Male': 'sum', 'Non-binary': 'sum'}).reset_index() #add up character count by year

# Plotting characters by gender representation per year
grouped_data.plot(x='Release', y=['Male', 'Female', 'Non-binary'], marker='o')
plt.xlabel('Release Year')
plt.ylabel('Number of Characters')
plt.title('Distribution of Video Game Characters by Gender Per Year')
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')

# Display the plot
plt.show()
```

The graph reveals male characters consistently dominate the video game landscape, with a higher average (35.7) compared to female(18.5) and non-binary(3) characters. Although there are yearly fluctuations in gender representation, some years show an increase in presence in comparison to a decline in male character presence, such as in 2021. Non-binary characters have a smaller presence overall, but there seems to be a notable increase in their representation in 2022.

# What is the gender distribution in for the Nintendo games?


```python
#Zooming in on specific example of gender bias which spans across publisher/developer
pub_filter =  VGame['Publisher'] == 'Nintendo'
nintendo_games = VGame[pub_filter].groupby('Title')['Gender'].value_counts().unstack()

#Plot of gender representation in Nintendo Games
nintendo_games.plot(kind='bar', figsize=(10, 6))
plt.title('Gender Representation Within Nintendo Games')
plt.ylabel('Count')
plt.xlabel(' ')
plt.legend(title='Gender')
plt.show()
```

Finding a publisher/developer that is representative of the entire industry is near impossible, but Nintendo’s impact on video games is greater than most. This graph shows the clear bias Nintendo’s top selling games have towards male characters. What isn’t clear from this graph is the sexism within the narratives. The only “Damsel in distress” would be Princess Peach within the Super Mario franchise, but the lack of substance behind the female characters is very evident. Princess Daisy and the low effort “girl version” characters are very seldom seen and almost never playable characters. The Pokemon series requires a bit of background knowledge but the primary “equal” representation lies in the fact that you can choose to play as a boy or girl. Beyond this, the gym leaders, side characters, and NPCs offer very little representation because they are nothing more than vessels to carry out turn-based battles. Zelda is an iconic game series but with a very strange twist, Link is the most significant character in the game, a boy.
